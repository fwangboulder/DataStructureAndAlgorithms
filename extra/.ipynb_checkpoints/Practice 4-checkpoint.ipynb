````python
#156 Binary Tree Upside Down
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    
    def upsideDownBinaryTree(self, root):
        """
        :type root: TreeNode
        :rtype: TreeNode
        """
        if not root:
            return root
        stack=[]
        while root:
            stack.append(root)
            root=root.left
        new_root=stack.pop()
        while len(stack)>0:
            parent=stack.pop()
            parent.left=None
            new_root.left=parent.right
            new_root.right=parent
            parent.right=None
            new_root=new_root.right
        return new_root
class TreeNode(object):
    def __init__(self,x):
        self.val=x
        self.left=None
        self.right=None
        
    
    
````

````python
#162
#method 1
class Solution(object):
    def findPeakElement(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        if len(nums)==1:
            return 0
        for i in range(0,len(nums)):
            if i==0 and i<len(nums)-1 and nums[i+1]<nums[i]:
                return i
            elif i==len(nums)-1 and nums[i-1]<nums[i]:
                return i
            elif nums[i+1]<nums[i] and nums[i-1]<nums[i]:
                return i
        return -1
#method 2
class Solution:
   
    def findPeakElement(self, nums):
        if not nums:
            return -1

        low, high = 0, len(nums) - 1
        while low <= high:
            if low == high:
                return low
            mid = (low + high) / 2
            if nums[mid] < nums[mid+1]:
                low = mid + 1
            else:
                high = mid

        return mid
````

````python
#159
class Solution(object):
    def lengthOfLongestSubstringTwoDistinct(self, s):
        """
        :type s: str
        :rtype: int
        """
        start=0
        d=set()
        count=0
        max_l=0
        i=0
        while i<len(s):
            d.add(s[i])
            count+=1
            i+=1
            if len(d)>2:
                count-=1
                max_l=max(max_l,count)
                start+=1
                i=start
                count=0
                d=set()
            max_l=max(max_l,count)
        return max_l
            
````

````python
#186
class Solution:
    # @param s, a list of 1 length strings, e.g., s = ['h','e','l','l','o']
    # @return nothing
    def reverseWords(self, s):
        self.reverse(s, 0, len(s) - 1)
    
        beg = 0
        for i in xrange(len(s)):
            if s[i] == ' ':
                self.reverse(s, beg, i-1)
                beg = i + 1
            elif i == len(s) -1:
                self.reverse(s, beg, i)

    def reverse(self, s, start, end):
        while start < end:
            s[start], s[end] = s[end], s[start]
            start += 1
            end -= 1 
````
