````python
class Solution(object):
    def diffWaysToCompute(self, input):
        """
        :type input: str
        :rtype: List[int]
        """
        if input.isdigit():
            return [int(input)]
        res = []
        for i in xrange(len(input)):
            if input[i] in "-+*":
                res1 = self.diffWaysToCompute(input[:i])
                res2 = self.diffWaysToCompute(input[i+1:])
                for j in res1:
                    for k in res2:
                        res.append(self.helper(j, k, input[i]))
        return res
    
    def helper(self, m, n, op):
        if op == "+":
            return m+n
        elif op == "-":
            return m-n
        else:
            return m*n
            
        
````

````python
class Solution(object):
    def searchInsert(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: int
        """
        if len(nums)==0:
            return 0
        l=0
        r=len(nums)-1
        while l<=r:
            mid=(l+r)//2
            if nums[mid]==target:
                return mid
            elif nums[mid]<target:
                l=mid+1
            else:
                r=mid-1
        return l
        
````

````python
class Solution(object):
    def lengthOfLongestSubstringTwoDistinct(self, s):
        """
        :type s: str
        :rtype: int
        """
        d={}
        start=0
        max_l=0
        for index,i in enumerate(s):
            d[i]=d.get(i,0)+1
            while len(d)>2:
                max_l=max(max_l,index-start)
                d[s[start]]-=1
                if d[s[start]]==0:
                    d.pop(s[start])
                start+=1
            max_l=max(max_l,index-start+1)
        return max_l
````

````python
class Solution(object):
    def lengthOfLongestSubstringKDistinct(self, s, k):
        """
        :type s: str
        :type k: int
        :rtype: int
        """
        d={}
        start=0
        i=0
        max_l=0
        while i<len(s):
            d[s[i]]=d.get(s[i],0)+1
            while len(d)>k:
                max_l=max(max_l,i-start)
                d[s[start]]-=1
                if d[s[start]]==0:
                    d.pop(s[start])
                start+=1
            max_l=max(max_l,i-start+1)
            i+=1
        return max_l
        
````
